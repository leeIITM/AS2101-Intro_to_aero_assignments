```python
array_l=[np.array(i) for i in l]
print(type(array_l[0]))
x1_1t=array_l[0].T
print(x1_1t)
```

```python
class linreg:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def findbeta(x,y):
    x1_1t=x.T
    XT_X = np.matmul(x1_1t, x)
    #print(XT_X)

    XT_y = np.matmul(x1_1t, y)
    #print(XT_y)

    betas = np.matmul(np.linalg.inv(XT_X), XT_y)
    return betas
  def plot_graph(x,y,betas):
    plt.figure(figsize=(8, 6))
    plotx=[]
    for i in range(len(x)):
      plotx.append(x[i][1])
    plotx = np.array(plotx)
    plt.scatter(plotx, y, color='red', label='Data Points')
    plt.plot(plotx, betas[0] + betas[1] * plotx, color='blue', label='Regression Line')
```
